###### loc iloc ix 介紹
loc ——通过行列的  标签  索引行数据

iloc——通过行列的  号索  引行数据

ix——通过行标签或者行号索引行数据

```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
csv_data_path = os.path.join('.\data', 'application_test.csv')
print('Path of read in data: %s' %(csv_data_path))
csv_data = pd.read_csv(csv_data_path)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
#print(csv_data.columns.size)
print(csv_data.iloc[0,:].size)
print(csv_data.iloc[:,0].size)
print(csv_data.shape)
```

#### 列出所有欄位

```python
print(csv_data)
```

#### 截取部分資料

```python
#print(csv_data([0:3]))
print("=====================================")
print(csv_data.iloc[0][0])
print("=====================================")
print(csv_data.iloc[3][3])
print("=====================================")
print(csv_data.iloc[3])
```
